```python
import numpy as np
import pandas as pd
```

```python
customer_transactions = pd.read_excel('customer_transaction_behavior.xlsx')
customer_transactions
```

```python
customer_transactions.info()
```

```python
# Trial period if from February 2019 to April 2019
# https://stackoverflow.com/questions/29370057/select-dataframe-rows-between-two-dates
mask = (customer_transactions['DATE'] > '2019-02-01') & (customer_transactions['DATE'] <= '2019-04-30')
trial_dates = customer_transactions.loc[mask]
trial_dates
```

```python
trial_period = trial_dates[['DATE', 'STORE_NBR', 'LYLTY_CARD_NBR', 'PROD_QTY', 'TOT_SALES', 'PROD_SIZE', 'BRAND_NAME', 'LIFESTAGE', 'PREMIUM_CUSTOMER']].copy()
trial_period
```

```python
# Performance Stores
performance_stores = trial_period[(trial_period['STORE_NBR'] == 77) | (trial_period['STORE_NBR'] == 86) | (trial_period['STORE_NBR'] == 88)]
performance_stores.groupby(['STORE_NBR']).agg({'PROD_QTY': ['mean'],'TOT_SALES': ['mean']})
```

```python
# Look for stores with similar performance in prod_qty and sales
# Store with sales more than 1.5 but less than 2 in mean
trial_dates[(trial_dates['TOT_SALES'] >= 8) & (trial_dates['PROD_QTY'] <= 9)]
```

```python
trial_dates.groupby(['STORE_NBR'])[['TOT_SALES']].mean().sort_values(by='TOT_SALES', ascending=False).to_string()
```

```python
# For store 77 : Store 53
# For store 86 : Store 97
# For store 88 : Store 153
control_stores = trial_period[(trial_period['STORE_NBR'] == 18) | (trial_period['STORE_NBR'] == 97) | (trial_period['STORE_NBR'] == 153)]
control_stores.groupby(['STORE_NBR']).agg({'PROD_QTY': ['mean'],'TOT_SALES': ['mean']})
```

```python
performance_stores
```

```python
performance_and_control = performance_stores.append(control_stores, ignore_index=True)
performance_and_control
```

```python
performance_and_control.groupby(['STORE_NBR']).agg({'PROD_QTY': 'mean', 'TOT_SALES': 'sum'})
```

```python
performance_and_control.groupby(['STORE_NBR']).agg({'PROD_QTY': 'mean', 'TOT_SALES': 'sum'})
```
